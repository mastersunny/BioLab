```python
import random
def dna_random_base():
    l = ['A','T','G','C']
    return l[random.randint(0,3)]

dna_random_base()
```

```python
import random
def rna_random_base():
    l = ['A','U','G','C']
    return l[random.randint(0,3)]

rna_random_base()
```

```python
def random_base(flag=False):
    if flag:
        return rna_random_base()
    else:
        return dna_random_base()

random_base(True)
```

```python
def random_dna_seq(l = 100):
    s = ""
    for i in range (l):
        s += dna_random_base()
    return s

random_dna_seq()
```

```python
def dna_monomer_freq(seq,base):
    return seq.count(base)*100/len(seq)
    
dna_monomer_freq('AATGCAA','A')   
```

```python
import textwrap

def save_fasta(filename,seq,id,desc):
    filename+='.fasta'
    with open(filename,'w') as f:
        f.write('>'+id+' <'+desc+'>\n')
        text = textwrap.wrap(seq,80)
        for x in text:
            f.write(x+'\n');
       
    print('file saved')
    
save_fasta('bio',random_dna_seq(1000),'2011331033','this is testing')
```

```python
def open_fasta(filename):
    with open(filename) as f:
        f.readline()
        lines = f.readlines()
        s=''
        for x in lines:
            s+=x
        return s
    
print(open_fasta('bio.fasta'))
```

```python
def transcribe(seq):
    return seq.replace('T','U')

transcribe('ATTTTGCTTTT')
```

```python
def complement(seq):
    dict = {'A':'T','T':'A','G':'C','C':'G'}
    s = ""
    for i in seq:
        s += dict[i]
    return s

complement("ATTAG")
```

```python
def rev_complement(seq):
    dict = {'A':'T','T':'A','G':'C','C':'G'}
    s = ""
    for i in seq:
        s += dict[i]
    return s[::-1]

rev_complement("ATTAG")
```

```python
def PatternCount(text,pattern):
    count=0
    n = len(text)
    k = len(pattern)
    for i in range(n-k+1):
        if pattern == text[i:k+i]:
            count+=1
    return count

print(PatternCount('GCGCG','GCG'))
```

```python
def MostFreq(text,k):
    n = len(text)
    dict ={}
    for i in range(n-k+1):
        pattern = text[i:k+i]
        if  pattern in dict:
            dict[pattern]+=1
        else:
            dict[pattern]=1
    max_val = max(dict.values())
    for key,value in dict.items():
        if value==max_val:
            print(key,end=" ")
        

MostFreq('ACGTTGCATGTCGCATGATGCATGAGAGCT',4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def skew(text):
    n = len(text)
    l = [0]*n
    for i in range(n):
        l[i] = text[:i].count('G')-text[:i].count('C')
    plt.plot(l)
    plt.ylabel("#G - #C")
    plt.xlabel('position')
    
    for i,val in enumerate(l):
        if val== min(l):
            print(i,end=" ")
    
skew("CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG")
```

```python
import operator

def hammingDist(seq1,seq2):
    return sum(map(operator.ne,seq1,seq2))

hammingDist("GGGCCGTTGGT","GGACCGTTGAC")
```

```python
def approx(pattern,text,d):
    n = len(text)
    k = len(pattern)
    for i in range(n-k+1):
        if hammingDist(text[i:k+i],pattern)<=d:
            print(i,end=" ")

approx("ATTCTGGA","CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC",3)
```

```python
import itertools

def getCombination(k):
    patterns = []
    for combination in itertools.product(['A','C','G','T'],repeat=k):
        patterns.append(''.join(map(str,combination)))
    return patterns
```

```python
import math

def patterntonumber(pattern):
    m = {'A':0,'C':1,'G':2,'T':3}
    
    sum_val = 0
    val = len(pattern)-1
    for i in pattern:
        sum_val += (math.pow(4,val)*m[i])
        val-=1
    print(int(sum_val))

patterntonumber("TAATGGTCACGTGGCATGTCTGG")
```

```python
def numbertopatttern(number,k):
    m = {0:'A',1:'C',2:'G',3:'T'}
    l = []
    while(number>0):
        l.append(m[number%4])
        number = number//4
    l.append(m[number%4])
    
    return ''.join(l)[k-1::-1]
    
        
numbertopatttern(5089,11)
```

```python
def number2pattern(number,k):
    N2P = {0:'A',1:'C',2:'G',3:'T'}
    res = ''
    while number > 0:
        res+=N2P[number%4];
        number //= 4
    res+=N2P[number]
    return (res[k-1::-1])
number2pattern(7356,9)
```
